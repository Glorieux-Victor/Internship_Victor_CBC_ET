```python
from gwpy.timeseries import TimeSeries
import matplotlib.pyplot as plt
import numpy as np

import sys
sys.path.append('../src')
from get_data import convert_data_dict_to_frequency_series_dict
from likelihood import MDCGaussianNoise
```

```python
tsd1 = TimeSeries.read('../data/loudest_BBH/loudest_E1.gwf', channel='E1:STRAIN')
tsd2 = TimeSeries.read('../data/loudest_BBH/loudest_E2.gwf', channel='E2:STRAIN')
tsd3 = TimeSeries.read('../data/loudest_BBH/loudest_E3.gwf', channel='E3:STRAIN')
```

```python
tsd = {'E1': tsd1, 'E2': tsd2, 'E3': tsd3}
fft, psd = convert_data_dict_to_frequency_series_dict(tsd)
```

```python
plt.loglog(fft['E1'].get_sample_frequencies(), np.abs(fft['E1'])**2)
plt.xlim(4, 1024)
```

```python
model = MDCGaussianNoise(fft, psd)
```

```python
model.maximize(max_iterations=1000)
```

```python
signal_reconstructed = model.reconstruct_signal()
```

```python
tsd_r = signal_reconstructed['E1'].to_timeseries()
tsd_r = tsd_r.cyclic_time_shift(19)
tsd = tsd1.bandpass(4,1024)
plt.plot(tsd_r.get_sample_times(), tsd_r)
plt.plot(tsd.times.value - 10016204, tsd, alpha=0.7)
```
